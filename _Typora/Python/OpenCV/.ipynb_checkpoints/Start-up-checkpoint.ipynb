# python+opencv图像处理

## 图像的导入和导出

```python
import cv2
```

```python
def get_image_info(image): # 定义加载图像信息函数
    print(type(image)) # 读取图像类别
    print(image.shape) # 加载图像的长宽和通道数
    print(image.size) # 图像的大小
    print(image.dtype) # 字节位数占多少
```

```python
img = cv2.imread('./imgs/1.png') # 读取图片
cv2.imshow('image', img) # 图片显示
get_image_info(img) # 调用图片信息函数
cv2.waitKey(0)  # 等待时间
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('./2.jpg', gray) # 写出到当前文件夹下
cv2.waitKey(0)
```

## 像素运算

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
def add_demo(m1, m2):
    '''
    定义像素相加的运算
    '''
    dst = cv.add(m1, m2) # 对像素进行加法运算
#     cv.imshow('add_demo', dst) # 显示相加后的图像
    plt.imshow(dst)
    
def subtract_demo(m1, m2):
    dst = cv.subtract(m1, m2)
#     cv.imshow('subtract_demo', dst)
    plt.imshow(dst)
```

```python
scr1 = cv.imread('imgs/1.png') # 读入图片
scr2 = cv.imread('imgs/3.png')
print(scr1.shape) # 打印图片大小
print(scr2.shape)
# cv.imshow('image1', scr1) # 显示图片
# cv.imshow('image2', scr2)
```

```python
subtract_demo(scr1, scr2)
```

```python
add_demo(scr1, scr2) # 调用函数
```
